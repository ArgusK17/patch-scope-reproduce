```python
from tools.inference import forward, modified_forward
from tools.utils import show_outputs, extract_hidden, extract_hidden_list, break2tokens
from tools.task_generation import gen_task
import numpy as np

from sklearn.linear_model import LinearRegression

def find_last(my_lists):
    element_to_find = '<0x0A>'
    index_list=[]
    for my_list in my_lists:
        if element_to_find in my_list:
            reversed_index = my_list[::-1].index(element_to_find)
            original_index = len(my_list) - 1 - reversed_index
            index_list.append(original_index)
        else:
            index_list.append(1)
    return np.array(index_list)

batch_size_train = 100
batch_count=5
PROMPTS_train=[]
for i in range(batch_count):
    task_train=gen_task(prompt_num=batch_size_train, example_num=5, seed=i)
    prompts_train=task_train.few_shot
    PROMPTS_train.append(prompts_train)

batch_size_test = 100
task = gen_task(prompt_num=batch_size_test, example_num=5, seed=0, mode="test")
prompts = task.few_shot
target_pos = find_last(break2tokens(tokenizer,prompts))

answers = []
token_lists = break2tokens(tokenizer, prompts)
for i, _ in enumerate(break2tokens(tokenizer,prompts)):
    answers.append(token_lists[i][target_pos[i]+1])

Record=np.zeros((40,40))

for target_layer_id in range(16):
    outputs = forward(model, tokenizer, prompts, with_hiddens=True)
    hiddens_target = extract_hidden_list(outputs, target_layer_id, token_pos_list=target_pos+1)

    # Evaluate basline rate
    result = id_lens(model, tokenizer, hiddens_target, target_layer_id)
    rate = (np.array(result)==np.array(answers)).sum()/batch_size_test

    for source_layer_id in range(target_layer_id,40):
        # Generate training data pair
        hiddens_source_list=[]
        hiddens_target_list=[]
        for prompts_train in PROMPTS_train:
            outputs_train=forward(model, tokenizer, prompts_train, with_hiddens=True)
            target_pos_train=find_last(break2tokens(tokenizer,prompts_train))
            hiddens_source=extract_hidden_list(outputs_train, source_layer_id, token_pos_list=target_pos_train+1)
            hiddens_source_list.append(hiddens_source)
            hiddens_target=extract_hidden_list(outputs_train, target_layer_id, token_pos_list=target_pos_train+1)
            hiddens_target_list.append(hiddens_target)
        X_train=torch.cat(hiddens_source_list,dim=0).cpu().numpy()
        Y_train=torch.cat(hiddens_target_list,dim=0).cpu().numpy()

        # Fit the linear model
        ln_reg = LinearRegression()
        ln_reg.fit(X_train, Y_train)

        # Generate testing data
        hiddens = extract_hidden_list(outputs, source_layer_id, token_pos_list=target_pos+1)

        # Inference with the linear model
        target_pred = ln_reg.predict(hiddens.cpu().numpy())
        hiddens_target_pred = torch.tensor(target_pred, dtype=torch.float16).to("cuda")

        # Use id_lens to evaluate the result
        result_rec = id_lens(model, tokenizer, hiddens_target_pred, target_layer_id)
        rate_rec=(np.array(result_rec)==np.array(answers)).sum()/batch_size_test
        print(f"source layer {source_layer_id:02d}, target layer {target_layer_id:02d}, original {rate*100}%, recover {rate_rec*100}%")
        Record[source_layer_id, target_layer_id]=rate_rec
```

```python
import numpy as np

def break2tokens(tokenizer, promtp: str):
    token_list=[]
    inputs = tokenizer(promtp, return_tensors="pt", padding=True)
    
    for s in inputs["input_ids"][0].tolist():
        output = tokenizer.decode(s)
        token_list.append(output)
    return token_list

print(break2tokens(tokenizer,'récit'))
```
